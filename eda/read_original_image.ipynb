## Thống kê cơ bản

```python
# thống kê số lượng file trong 1 thư mục
import os
def count_files_in_directory(directory):
    """Đếm số lượng file trong thư mục chỉ định."""
    file_count = 0
    for root, dirs, files in os.walk(directory):
        file_count += len(files)
    return file_count
```

```python
# lấy tên thư mục con trong thư mục extract_zip/Image_library
def list_subdirectories(directory):
    """Lấy danh sách tên các thư mục con trong thư mục chỉ định."""
    subdirectories = []
    for entry in os.scandir(directory):
        if entry.is_dir():
            subdirectories.append(entry.name)
    return subdirectories
```

```python
categories = list_subdirectories('extract_zip/Image_library')
len(categories)
```

```python
print(categories)
```

```python
# đếm số lượng file trong từng thư mục con
# mỗi thư mục con là 1 key, giá trị là số lượng ảnh trong thư mục đó
counts = {}
for category in categories:
    dir_path = os.path.join('extract_zip/Image_library', category)
    counts[category] = count_files_in_directory(dir_path)
```

```python
# các thông số thống kê cơ bản
total_categories = len(counts)
total_files = sum(counts.values())
average_files_per_category = total_files / total_categories if total_categories > 0 else 0

print(f"Tổng số loại cá: {total_categories}")
print(f"Tổng số ảnh: {total_files}")
print(f"Số ảnh trung bình mỗi loại: {average_files_per_category:.2f}")
```

```python
# Kiểm tra số lượng outliers bằng IQR method
Q1 = df['Image_Count'].quantile(0.25)
Q3 = df['Image_Count'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = df[(df['Image_Count'] < lower_bound) | (df['Image_Count'] > upper_bound)]
print(f"Số lượng outliers phát hiện được: {len(outliers)}")
print("lower_bound:", lower_bound)
print("upper_bound:", upper_bound)
print(outliers)
```

## Phân tích Pareto (80/20 Rule)

```python
# Tính phần trăm tích lũy
df_sorted = df.sort_values('Image_Count', ascending=False).reset_index(drop=True)
df_sorted['Cumulative_Images'] = df_sorted['Image_Count'].cumsum()
df_sorted['Cumulative_Percentage'] = (df_sorted['Cumulative_Images'] / df_sorted['Image_Count'].sum()) * 100
df_sorted['Family_Percentage'] = ((df_sorted.index + 1) / len(df_sorted)) * 100

# Tìm điểm 80%
pareto_80_idx = df_sorted[df_sorted['Cumulative_Percentage'] >= 80].index[0]
pareto_80_families = pareto_80_idx + 1
pareto_80_pct = (pareto_80_families / len(df_sorted)) * 100

print("=== PHÂN TÍCH PARETO (80/20) ===")
print(f"Tổng số ảnh: {df_sorted['Image_Count'].sum():,}")
print(f"Tổng số họ cá: {len(df_sorted)}")
print(f"\n80% số lượng ảnh đến từ {pareto_80_families} họ cá ({pareto_80_pct:.1f}% tổng số họ)")
print(f"20% họ cá đầu chiếm {df_sorted.head(int(len(df_sorted)*0.2))['Image_Count'].sum() / df_sorted['Image_Count'].sum() * 100:.1f}% tổng số ảnh")

# Vẽ Pareto Chart
fig, ax1 = plt.subplots(figsize=(16, 8))

# Bar chart
x_pos = np.arange(min(75, len(df_sorted)))
bars = ax1.bar(x_pos, df_sorted['Image_Count'].head(75), color='#3498db', alpha=0.7, edgecolor='black')
ax1.set_xlabel('Thứ tự họ cá (Top 75)', fontweight='bold', fontsize=12)
ax1.set_ylabel('Số lượng ảnh', fontweight='bold', fontsize=12, color='#3498db')
ax1.tick_params(axis='y', labelcolor='#3498db')

# Line chart
ax2 = ax1.twinx()
ax2.plot(x_pos, df_sorted['Cumulative_Percentage'].head(75), color='#e74c3c', marker='o', 
         linewidth=2, markersize=4, label='Tích lũy %')
ax2.set_ylabel('Phần trăm tích lũy (%)', fontweight='bold', fontsize=12, color='#e74c3c')
ax2.tick_params(axis='y', labelcolor='#e74c3c')
ax2.axhline(80, color='green', linestyle='--', linewidth=2, label='80% threshold')
ax2.set_ylim([0, 105])

# Title and grid
plt.title('Biểu đồ Pareto - Top 75 họ cá', fontweight='bold', fontsize=14)
ax1.grid(True, alpha=0.3, axis='y')
ax2.legend(loc='center right')

plt.tight_layout()
plt.show()

print(f"\n=== TOP {pareto_80_families} HỌ CÁ (CHIẾM 80% DỮ LIỆU) ===")
print(df_sorted.head(pareto_80_families)[['Family', 'Image_Count', 'Cumulative_Percentage']].to_string(index=False))
```

## Phân tích mẫu ảnh (Sample Analysis)

```python
import os
from PIL import Image
import random

# Lấy 9 họ cá ngẫu nhiên từ các nhóm khác nhau
samples = {
    'Nhiều ảnh': df[df['Image_Count'] >= 1000].sample(min(3, len(df[df['Image_Count'] >= 1000]))),
    'Trung bình': df[(df['Image_Count'] >= 100) & (df['Image_Count'] < 1000)].sample(min(3, len(df[(df['Image_Count'] >= 100) & (df['Image_Count'] < 1000)]))),
    'Ít ảnh': df[df['Image_Count'] < 100].sample(min(3, len(df[df['Image_Count'] < 100])))
}

fig, axes = plt.subplots(3, 3, figsize=(15, 15))
fig.suptitle('MẪU ẢNH TỪ CÁC HỌ CÁ KHÁC NHAU', fontsize=16, fontweight='bold')

idx = 0
for group_name, sample_df in samples.items():
    for _, row in sample_df.iterrows():
        family_name = row['Family']
        img_count = row['Image_Count']
        
        # Đường dẫn đến thư mục họ cá
        family_path = os.path.join('extract_zip/Image_Library', family_name)
        
        if os.path.exists(family_path):
            # Lấy danh sách ảnh
            images = [f for f in os.listdir(family_path) if f.endswith('.jpg')]
            
            if images:
                # Chọn ngẫu nhiên 1 ảnh
                random_img = random.choice(images)
                img_path = os.path.join(family_path, random_img)
                
                try:
                    # Đọc và hiển thị ảnh
                    img = Image.open(img_path)
                    row_idx = idx // 3
                    col_idx = idx % 3
                    axes[row_idx, col_idx].imshow(img)
                    axes[row_idx, col_idx].set_title(f'{family_name}\n({img_count} ảnh) - {group_name}', 
                                                      fontsize=10, fontweight='bold')
                    axes[row_idx, col_idx].axis('off')
                    idx += 1
                except Exception as e:
                    print(f"Lỗi khi đọc ảnh {img_path}: {e}")

plt.tight_layout()
plt.show()
```
